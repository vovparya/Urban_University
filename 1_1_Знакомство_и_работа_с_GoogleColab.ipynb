````python
import numpy as np
import matplotlib.pyplot as plt

# Генерация случайных данных
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Построение графика
plt.plot(x, y)
plt.title('Пример графика в Google Colab')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
````
