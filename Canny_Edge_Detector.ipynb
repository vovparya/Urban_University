### Canny Edge Detector

Canny Edge Detector — это популярный алгоритм для обнаружения границ в изображениях, разработанный Джоном Кэнни в 1986 году. Он предназначен для выявления важных границ и контуров объектов на изображениях, сводя к минимуму уровень шума и ложные срабатывания. Алгоритм состоит из нескольких этапов, каждый из которых выполняет определённую задачу для достижения точных и надежных результатов.

### Этапы работы алгоритма Canny:

1. **Сглаживание изображения:**
   - Прежде чем приступать к обнаружению границ, изображение сглаживается с помощью гауссового фильтра для удаления шума. Это помогает избежать ложных границ, вызванных шумом. Гауссов фильтр обычно представлен следующим образом:

2. **Вычисление градиентов:**
   - Далее, для каждого пикселя изображения вычисляются градиенты интенсивности в направлении X и Y. Это достигается с помощью операторов Собела:

3. **Подавление немаксимумов (Non-maximum suppression):**
   - На этом этапе удаляются неважные границы, которые не являются локальными максимумами в направлении градиента. Это помогает тонко выделить настоящие границы. Пиксели, которые не являются локальными максимумами, устанавливаются в ноль.

4. **Двойная пороговая обработка (Double thresholding):**
   - Используются два порога: высокий и низкий. Пиксели с градиентами выше высокого порога считаются "сильными" границами, а пиксели с градиентами между низким и высоким порогами считаются "слабыми" границами. Пиксели с градиентами ниже низкого порога отбрасываются.

5. **Гистерезис (Hysteresis):**
   - Этот этап используется для определения окончательных границ. Слабые границы включаются в окончательный набор границ, только если они связаны с сильными границами. Это позволяет исключить слабые шумовые границы и сохранить только значимые контуры.

### Преимущества и недостатки:

#### Преимущества:
- **Точность:** Алгоритм эффективно находит границы и минимизирует ложные срабатывания.
- **Шумоустойчивость:** За счет предварительного сглаживания изображение очищается от шума.
- **Сохранение важных деталей:** Использование двойной пороговой обработки и гистерезиса позволяет сохранить важные границы.

#### Недостатки:
- **Чувствительность к параметрам:** Результаты сильно зависят от выбора параметров (размера гауссового фильтра, значений порогов).
- **Высокая вычислительная сложность:** По сравнению с простыми методами, Canny требует больше вычислительных ресурсов.

### Пример использования:

Ниже приведен пример применения алгоритма Canny Edge Detector с использованием OpenCV на Python

Этот код загружает изображение, применяет к нему Canny Edge Detector с пороговыми значениями 100 и 200, а затем отображает оригинальное изображение и результаты обработки.



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Загрузка изображения в оттенках серого
image = cv2.imread('/content/color_text.jpg', cv2.IMREAD_GRAYSCALE)

# Применение Canny Edge Detector
edges = cv2.Canny(image, 100, 200)

# Отображение результатов
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.subplot(1, 2, 2)
plt.title('Canny Edges')
plt.imshow(edges, cmap='gray')
plt.show()
```

Canny Edge Detector остается одним из самых мощных и часто используемых методов для обнаружения границ в изображениях благодаря своей точности и надежности.
